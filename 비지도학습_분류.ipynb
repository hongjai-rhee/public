# 원하는 클러스터의 수를 정해서, K-means 분류 알고리즘 실행  

```
# Number of clusters
n_clust=3
kmeans = KMeans(n_clusters=n_clust)
# Fitting the input data
kmeans = kmeans.fit(z)
# Getting the cluster labels
labels = kmeans.predict(z)
# Centroid values
centroids = kmeans.cluster_centers_
print(centroids)
print(labels)
```
